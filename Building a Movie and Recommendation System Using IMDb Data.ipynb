```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/sudey/Downloads/users.data",sep="\t",names=["user_id","item_id","rating","timestamp"])
```

```python
data.head()
```

```python
movies=pd.read_csv("C:/Users/sudey/Downloads/movie_id_titles.csv")
```

```python
movies.head()
```

```python
data=pd.merge(data,movies,on="item_id")
```

```python
data.head()
```

```python
table=data.pivot_table(index="user_id",columns="title",values="rating")
table.head()
```

```python
starwars=table["Star Wars (1977)"]
starwars.head()
```

```python
correlation = table.corrwith(starwars)
```

```python
correlation
```

```python
correlation_Df=pd.DataFrame(correlation,columns=["Correlation"])
correlation_Df.dropna(inplace=True)
correlation_Df.head()
```

```python
correlation_Df.sort_values("Correlation",ascending=False).head(15)
```
